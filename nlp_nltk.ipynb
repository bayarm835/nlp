```python
#Open the text file : 
text_file = open("texte_nlp.txt")
```

```python
text = text_file.read()
```

```python
print(type(text))
print("\n")
```

```python
print(text)
print("\n")
```

```python
print(len(text))
```

```python
import nltk
from nltk import sent_tokenize
from nltk import word_tokenize
```

```python
#tokenize the text by sentences
sentences = sent_tokenize(text)
```

```python
print(len(sentences))
```

```python
sentences
```

```python
#Tokenize the text with words
words = word_tokenize(text)
```

```python
#How many words are there ?
print(len(words))
```

```python
print(words)
```

```python
from nltk import FreqDist

#Trouver la répetition des mots
fdist = FreqDist(words)

#Afficher les 10 mots les plus utilisées
fdist.most_common(10)
```

```python
#Afficher avec un graph Plot
import matplotlib.pyplot as Plot
fdist.plot(20)
```

```python
#Une liste vide pour stocker les mots
words_no_punc = []

#Enlever les points
for w in words:
    if w.isalpha():
        words_no_punc.append(w.lower())

#Afficher le smots sans les points
print(words_no_punc)

print("\n")

#Et afficher la longueur
print(len(words_no_punc))
```

```python
#Afficher avec un graph Plot

#Trouver la répetition des mots
fdist_no_punc = FreqDist(words_no_punc)

fdist_no_punc.plot(20)
```

```python
# Notice that we still have many words that are not very useful in the 
# analysis of our text file sample, such as “and,” “but,” “so,” and others. 
# Next, we need to remove coordinating conjunctions.
```

```python
from nltk.corpus import stopwords
nltk.download('stopwords')

#La liste avec les stopwords
stopwords = stopwords.words("english")
print(stopwords)
```

```python
#Une liste vide pour stocker les mots sans les stopwords
clean_words = []

for w in words_no_punc:
    if w not in stopwords:
        clean_words.append(w)

print(clean_words)
print("\n")
print(len(clean_words))
```

```python
fdist_clean = FreqDist(clean_words)

fdist_clean.most_common(10)
```

```python
fdist_clean.plot(10)
```

```python
# As shown above, the final graph has many useful words that help us 
# understand what our sample data is about, showing how essential it is 
# to perform data cleaning on NLP.
```
